```python
import pandas as pd
from datetime import datetime as dt
import json as js
```

```python
with open ("fotmobSeason2010_2011.json") as ft:
    jsDat = js.load(ft)
```

```python
class FotMob2010_2011:    
    def Details(self):
        info = jsDat['details']
        Id = info['id']
        Type = info['type']
        Name = info['name']
        SSeas = info['selectedSeason']
        LSeas = info['latestSeason']
        SName = info['shortName']
        Cntry = info['country']
        print(f"{Id}\n{Type}\n{Name}\n{SSeas}\n{LSeas}\n{SName}\n{Cntry}\n")

    def Table(self):
        Tabl = jsDat['table']
        for i in Tabl:
            CCode = i['data']['ccode']
            LgId = i['data']["leagueId"]
            lgnd = i['data']['legend']
            TblFltTyp = i['data']['tableFilterTypes']
            SSeas = i['data']['selectedSeason']
            Tble = i['data']['table']
            for i in lgnd:
                Tkey = i['tKey']
                Indcs = i['indices']
                print(f"\n{Tkey}\n{Indcs}")
            print(f"{CCode}\n{LgId}\n{TblFltTyp}\n{SSeas}\n")
            for i in Tble.keys():
                print(f"\n{i.upper()}")
                for i in Tble[i]:
                    Nme = i['name']
                    ShNme = i['shortName'] 
                    Id =  i['id']
                    PgUrl = i['pageUrl']
                    Plyd = i['played']
                    Win = i[ 'wins'] 
                    Drw = i['draws']
                    Loss = i['losses']
                    ScrStr = i['scoresStr']
                    GoalDiff = i['goalConDiff'] 
                    pts = i['pts']
                    Indx = i['idx']
                    print(f"\n{Nme}\n{ShNme}\n{Id}\n{PgUrl}\n{Plyd}\n{Win}\n{Drw}\n{Loss}\n{ScrStr}\n{GoalDiff}\n{pts}\n{Indx}")

    def tableTeamForm(self):
        Tabl = jsDat['table']
        for i in Tabl:
            Tf = i['teamForm']
            for i in Tf.keys():
                kk = Tf[i]
                for i in kk:
                    Res = i['result']
                    ResStrng = i['resultString'] 
                    LnkTMatch = i['linkToMatch'] 
                    TeamPgUrl = i['teamPageUrl']
                    Scr = i['score']
                    ToolTip = i['tooltipText']
                    UtcTm = ToolTip['utcTime']
                    HmTm = ToolTip['homeTeam']
                    HmTmId = ToolTip['homeTeamId']
                    HmScr = ToolTip['homeScore']
                    AwyTm = ToolTip['awayTeam']
                    AwyTmId = ToolTip['awayTeamId']
                    AwyScr = ToolTip['awayScore']
                    print(f"{Res}\n{ResStrng}\n{LnkTMatch}\n{TeamPgUrl}\n{Scr}\n{UtcTm}\n{HmTm}\n{HmTmId}\n{HmScr}\n{AwyTm}\n{AwyTmId}\n{AwyScr}\n")

    def Overview(self):
        Tabl = jsDat['overview']['table']
        for i in Tabl:
            CCode = i['data']['ccode']
            LgId = i['data']["leagueId"]
            lgnd = i['data']['legend']
            TblFltTyp = i['data']['tableFilterTypes']
            SSeas = i['data']['selectedSeason']
            Tble = i['data']['table']
            for i in lgnd:
                Tkey = i['tKey']
                Indcs = i['indices']
                print(f"\n{Tkey}\n{Indcs}")
            print(f"{CCode}\n{LgId}\n{TblFltTyp}\n{SSeas}\n")
            for i in Tble.keys():
                print(f"\n{i.upper()}")
                for i in Tble[i]:
                    Nme = i['name']
                    ShNme = i['shortName'] 
                    Id =  i['id']
                    PgUrl = i['pageUrl']
                    Plyd = i['played']
                    Win = i[ 'wins'] 
                    Drw = i['draws']
                    Loss = i['losses']
                    ScrStr = i['scoresStr']
                    GoalDiff = i['goalConDiff'] 
                    pts = i['pts']
                    Indx = i['idx']
                    print(f"\n{Nme}\n{ShNme}\n{Id}\n{PgUrl}\n{Plyd}\n{Win}\n{Drw}\n{Loss}\n{ScrStr}\n{GoalDiff}\n{pts}\n{Indx}")
                    
    def OverviewTeamForm(self):
        Tabl = jsDat['overview']['table']
        for i in Tabl:
            Tf = i['teamForm']
            for i in Tf.keys():
                kk = Tf[i]
                print(i)
                for i in kk:
                    Res = i['result']
                    ResStrng = i['resultString'] 
                    LnkTMatch = i['linkToMatch'] 
                    TeamPgUrl = i['teamPageUrl']
                    Scr = i['score']
                    ToolTip = i['tooltipText']
                    UtcTm = ToolTip['utcTime']
                    HmTm = ToolTip['homeTeam']
                    HmTmId = ToolTip['homeTeamId']
                    HmScr = ToolTip['homeScore']
                    AwyTm = ToolTip['awayTeam']
                    AwyTmId = ToolTip['awayTeamId']
                    AwyScr = ToolTip['awayScore']
                    print(f"{Res}\n{ResStrng}\n{LnkTMatch}\n{TeamPgUrl}\n{Scr}\n{UtcTm}\n{HmTm}\n{HmTmId}\n{HmScr}\n{AwyTm}\n{AwyTmId}\n{AwyScr}\n")

    def leagueOverviewMatches(self):
        lgOvrMat = jsDat['overview']['leagueOverviewMatches']
        for i in lgOvrMat:
            pgUrl = i['pageUrl']        
            Id = i['id']
            Res = i['status']['reason']['short']  
            Scr = i['status']['scoreStr']
            UtcTm = i['status']['utcTime']  
            hna = i['home']['name']
            hSn = i['home']['shortName']
            hId = i['home']['id']
            ana = i['away']['name']
            aSn = i['away']['shortName']
            aId = i['away']['id']
            print(f"{hna}\n{hSn}\n{hId}\n{ana}\n{aSn}\n{aId}\n{Res}\n{Scr}\n{pgUrl}\n{Id}\n{UtcTm}\n")

    def AllMatches(self):
        for i in jsDat['matches']['allMatches']:
            pgUrl = i['pageUrl']        
            Id = i['id']
            Res = i['status']['reason']['short']  
            Scr = i['status']['scoreStr']
            UtcTm = i['status']['utcTime']  
            hna = i['home']['name']
            hSn = i['home']['shortName']
            hId = i['home']['id']
            ana = i['away']['name']
            aSn = i['away']['shortName']
            aId = i['away']['id']
            print(f"{hna}\n{hSn}\n{hId}\n{ana}\n{aSn}\n{aId}\n{Res}\n{Scr}\n{pgUrl}\n{Id}\n{UtcTm}\n")

            
```
