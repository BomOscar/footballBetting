```python
import pandas as pd
import json as js
from datetime import datetime 
```

```python
with open("fotmobSeason2010_2011.json") as fm:
    data = js.load(fm)
type(data)
    
```

```python
details = []
for i in data['details'].keys():
    new = data['details'][i]
    details.append(f'{i}\n{new}\n')
```

```python
detail = []
for i in details:
    detail.append(i.splitlines())
```

```python
dd = pd.DataFrame(detail)
```

```python
dd.to_csv("details.csv")
```

```python
tables = []
t = []
rr = []
for i in data['table'][0].keys():
    inf = data['table'][0][i]
    for i in inf['table'].keys():
        morInf = inf['table'][i]
        for i in morInf:
            tables.append(i.items())
            t.append(i.keys())
            rr.append(i)
           
```

```python
rr = []
for i in data['matches'].keys():
    inf = data['matches'][i]
    dataa = data['matches']['allMatches']
    for i in dataa:
        pageUrl = i['pageUrl']
        id = i['id']
        rr.append({
            'id': id,
            'pageUrl': pageUrl,
        })
        homeName = i['home']
        for i in homeName.keys():
            inf = homeName[i]
            rr[-1][i] = inf

df = pd.DataFrame(rr)
print(df)
```
